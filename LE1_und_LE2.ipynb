```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download(handle="nicholasjhana/energy-consumption-generation-prices-and-weather")

print("Path to dataset files:", path)

path_energy = path + "/energy_dataset.csv"
path_weather = path + "/weather_features.csv"

# Load the datasets
import pandas as pd

energy_df = pd.read_csv(path_energy)
weather_df = pd.read_csv(path_weather)
```

```python
energy_df['time'] = pd.to_datetime(energy_df['time'], utc=True)
energy_df = energy_df.set_index('time')

weather_df['dt_iso'] = pd.to_datetime(weather_df['dt_iso'], utc=True)
weather_df = weather_df.set_index('dt_iso')
```

```python
weather_df_reset = weather_df.reset_index()

# Columns to check for differences
all_cols = weather_df_reset.columns.tolist()
weather_specific_cols = ['weather_id', 'weather_main', 'weather_description', 'weather_icon']
other_cols = [c for c in all_cols if c not in weather_specific_cols and c not in ['dt_iso', 'city_name']]

# Get duplicates for one city as example
city = 'Madrid'
city_data = weather_df_reset[weather_df_reset['city_name'] == city].copy()

# Find duplicate timestamps
duplicate_times = city_data[city_data.duplicated(subset=['dt_iso'], keep=False)].sort_values('dt_iso')
print(f"\nNumber of rows involved in duplicates for {city}: {len(duplicate_times)}")

# Show a few examples of duplicate pairs
print(f"\n--- Example duplicate pairs for {city} ---")
sample_times = duplicate_times['dt_iso'].unique()[:3]

for t in sample_times:
    print(f"\nTimestamp: {t}")
    rows = duplicate_times[duplicate_times['dt_iso'] == t]
    print(rows[['weather_id', 'weather_main', 'weather_description', 'weather_icon']].to_string())
    
    # Check if other columns are identical
    other_values = rows[other_cols]
    if other_values.drop_duplicates().shape[0] == 1:
        print("  â†’ All non-weather columns are IDENTICAL")
    else:
        print("  â†’ Some non-weather columns DIFFER:")
```

```python
# OpenWeatherMap weather condition codes reference:
# 2xx: Thunderstorm
# 3xx: Drizzle
# 5xx: Rain
# 6xx: Snow
# 7xx: Atmosphere (mist, fog, etc.)
# 800: Clear
# 80x: Clouds

print("\n--- Weather ID distribution ---")
weather_id_counts = weather_df['weather_id'].value_counts().sort_index()
print(weather_id_counts)

print("\n--- Weather main categories ---")
print(weather_df['weather_main'].value_counts())

# Check how many times each timestamp appears with multiple weather conditions
print("\n--- Distribution of duplicate counts per timestamp per city ---")
weather_df_reset = weather_df.reset_index()
dup_counts = weather_df_reset.groupby(['city_name', 'dt_iso']).size()
print(dup_counts.value_counts().sort_index())

print("\n--- Max duplicates per timestamp per city ---")
print(f"Max occurrences of same timestamp for a city: {dup_counts.max()}")
```

```python
weather_df_reset = weather_df.reset_index()
dup_counts = weather_df_reset.groupby(['city_name', 'dt_iso']).size().reset_index(name='count')

# Get timestamps with 3 or 4 conditions
multi_conditions = dup_counts[dup_counts['count'] >= 3]
print(f"\nTimestamps with 3+ weather conditions: {len(multi_conditions)}")

# Show examples
print("\n--- Examples of timestamps with 3+ conditions ---")
for _, row in multi_conditions.head(3).iterrows():
    city, ts, count = row['city_name'], row['dt_iso'], row['count']
    print(f"\n{city} at {ts} ({count} conditions):")
    subset = weather_df_reset[(weather_df_reset['city_name'] == city) & 
                              (weather_df_reset['dt_iso'] == ts)]
    print(subset[['weather_id', 'weather_main', 'weather_description']].to_string())

from collections import Counter

pair_counts = Counter()
for (city, ts), group in weather_df_reset.groupby(['city_name', 'dt_iso']):
    if len(group) > 1:
        weather_types = tuple(sorted(group['weather_main'].unique()))
        pair_counts[weather_types] += 1

print("\n--- Most common weather condition combinations ---")
for combo, count in pair_counts.most_common(15):
    print(f"{combo}: {count}")
```

```python
weather_types = ['rain', 'drizzle', 'mist', 'fog', 'thunderstorm', 'snow', 'clouds', 'clear', 'haze', 'dust']

# Create lowercase weather_main for matching
weather_df_reset['weather_main_lower'] = weather_df_reset['weather_main'].str.lower()

# Create binary columns for each weather type using vectorized str.contains
for wtype in weather_types:
    weather_df_reset[f'has_{wtype}'] = weather_df_reset['weather_main_lower'].str.contains(wtype, na=False)

# Aggregate binary columns
binary_cols = [f'has_{wtype}' for wtype in weather_types]
numeric_cols = ['temp', 'temp_min', 'temp_max', 'pressure', 'humidity', 
                'wind_speed', 'wind_deg', 'rain_1h', 'rain_3h', 'snow_3h', 'clouds_all']
categorical_cols = ['weather_id', 'weather_main', 'weather_description', 'weather_icon']

# Define aggregation strategy: max for binary (OR), first for others
agg_dict = {col: 'max' for col in binary_cols}
agg_dict.update({col: 'first' for col in numeric_cols if col in weather_df_reset.columns})
agg_dict.update({col: 'first' for col in categorical_cols if col in weather_df_reset.columns})

# Group and aggregate
weather_clean_binary = weather_df_reset.groupby(['dt_iso', 'city_name']).agg(agg_dict).reset_index()

# Clean up temporary column
if 'weather_main_lower' in weather_df_reset.columns:
    weather_df_reset.drop(columns=['weather_main_lower'], inplace=True)

print(f"Result: {len(weather_clean_binary)} rows with {len(weather_types)} binary columns")

# Show sample of binary features
print("\n--- Sample of binary weather features ---")
sample_cols = ['dt_iso', 'city_name'] + binary_cols[:5]
print(weather_clean_binary[weather_clean_binary[binary_cols].any(axis=1)][sample_cols].head(10).to_string())

# Verify no duplicates remain
dup_check = weather_clean_binary.groupby(['dt_iso', 'city_name']).size()
print(f"\nMax occurrences per (timestamp, city): {dup_check.max()} (should be 1)")
```

```python
# Check timestamp alignment between datasets
energy_timestamps = set(energy_df.index)
weather_timestamps_per_city = weather_clean_binary.groupby('city_name')['dt_iso'].apply(set)

print("\n--- Timestamp alignment check ---")
for city in weather_clean_binary['city_name'].unique():
    city_timestamps = weather_timestamps_per_city[city]
    overlap = len(energy_timestamps & city_timestamps)
    missing_in_weather = len(energy_timestamps - city_timestamps)
    extra_in_weather = len(city_timestamps - energy_timestamps)
    print(f"{city}: {overlap} matching, {missing_in_weather} missing in weather, {extra_in_weather} extra in weather")

weather_df_clean = weather_clean_binary.set_index('dt_iso')
print(f"\nCleaned weather dataset shape: {weather_df_clean.shape}")
print(f"Energy dataset shape: {energy_df.shape}")
```
